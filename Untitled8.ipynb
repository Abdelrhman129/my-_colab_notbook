```python
import pandas as pd

df = pd.read_csv("/content/Body Measurements _ original_CSV.csv")  # غيّر الاسم لو مختلف
df.head()
df.info()
df.describe()
df.columns  # عرض أسماء الأعمدة للتأكد
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# حمل ملف الإكسل
df = pd.read_csv("/content/Body Measurements _ original_CSV.csv")  # غيّر الاسم لو محتاج

# شيل أي صف فاضي
df.dropna(inplace=True)

# اختار الأعمدة اللي هنستخدمها في التحليل (غيّر حسب الموجود بالضبط)
X = df[["Age", "HeadCircumference","ShoulderWidth","TotalHeight","Gender"]]
# نعمل النموذج
kmeans = KMeans(n_clusters=3, random_state=42)
df["BodyTypeCluster"] = kmeans.fit_predict(X)

# عرض عدد الأفراد في كل مجموعة
print(df["BodyTypeCluster"].value_counts())
plt.scatter(df["HeadCircumference"], df["ShoulderWidth"], c=df["TotalHeight"])
plt.xlabel("HeadCircumference")
plt.ylabel("ShoulderWidth")
plt.title("TotalHeight")
plt.show()
def get_recommendations(cluster):
    if cluster == 0:
        return "تمارين زيادة عضل، نظام غذائي عالي السعرات"
    elif cluster == 1:
        return "تمارين متوازنة، نظام غذائي متوسط"
    else:
        return "تمارين حرق دهون، نظام غذائي قليل السعرات"

df["Recommendations"] = df["BodyTypeCluster"].apply(get_recommendations)
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df["HeadCircumference"], df["ShoulderWidth"], c=df["TotalHeight"], cmap="viridis")
plt.xlabel("HeadCircumference")
plt.ylabel("ShoulderWidth")
plt.title("TotalHeight")
plt.colorbar(label="Cluster")
plt.show()
df.to_csv("body_type_with_recommendations.csv", index=False)
age = int(input("السن: "))
head = float(input("محيط الرأس: "))
shoulder = float(input("عرض الكتف: "))
height = float(input("الطول الكلي: "))
gender = int(input("النوع (1 = ذكر, 2 = أنثى): "))

new_data = pd.DataFrame([{
    "Age": age,
    "HeadCircumference": head,
    "ShoulderWidth": shoulder,
    "TotalHeight": height,
    "Gender": gender
}])
# التنبؤ بالكلستر
cluster_label = kmeans.predict(new_data)[0]

# طباعة التوصية
recommendation = get_recommendations(cluster_label)

print(f"\n🔎 نوع الجسم المتوقع: {cluster_label}")
print(f"✅ التوصية: {recommendation}")
```
