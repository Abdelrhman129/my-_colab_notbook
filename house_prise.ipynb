```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import fetch_california_housing
```

```python
housing=fetch_california_housing()
```

```python
housing
```

```python
df=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
df
```

```python
df['Price']=housing.target
```

```python
df
```

```python
plt.scatter(df['MedInc'],df['Price'],color='blue',alpha=0.5)
plt.xlabel('Median Income')
plt.ylabel('House Price')
plt.title('Relation Between Income and House Price')
plt.show()
```

```python
plt.scatter(df['Longitude'],df['Price'],color='blue',alpha=0.5)
plt.xlabel('	Longitudee')
plt.ylabel('House Price')
plt.title('Relation Between Income and House Price')
plt.show()
```

```python
correlation_matrix=df.corr()
print("\ncorrelation with house Price:")
print(correlation_matrix["Price"].sort_values(ascending=False))
```

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
feature_to_explore=['MedInc','HouseAge','AveRooms','AveOccup','Population']
for feature in feature_to_explore:
    plt.figure(figsize=(10,6))
    plt.scatter(df[feature],df['Price'],alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(f'Relation Between {feature} and House Price')
    plt.show()
```

```python
selected_feature=[
"MedInc"	,"HouseAge"	,"AveBedrms"	,"AveOccup"	,"Latitude","Longitude",]
```

```python
x=df[selected_feature]
y=df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
modil=LinearRegression()
modil.fit(x_train,y_train)
```

```python
y_pred=modil.predict(x_test)
```

```python
y_pred
```

```python
#Evaluation model_perform
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
r2=r2_score(y_test,y_pred)
```
