```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set_theme(style="darkgrid")
```

```python
ciro_temp=[30,32,31,29,28,27,26]
alex_temp=[52,31,30,29,28,27,26]
sharm_temp=[32,31,30,29,28,27,26]
bnha_temp=[32,31,30,29,28,27,26]
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.plot(days,alex_temp,label="alex_temp",color="red",linewidth=2)
plt.plot(days,sharm_temp,label="sharm_temp",color="blue",linewidth=2)
plt.plot(days,ciro_temp,label="ciro_temp",color="green",linewidth=2)
plt.plot(days,bnha_temp,label="bnha_temp",color="yellow",linestyle="--")
plt.title("weekly Tempreture")
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.legend()
plt.show()
```

# numper5

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="day",data=tips,palette="viridis")
plt.title("count plot")
plt.show()
```

```python
sns.boxenplot(x="day",y="smoker",data=tips,palette="coolwarm")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,palette="coolwarm")
```

#numper6

```python
num_cpus=tips.corr(numeric_only=True)
plt.figure(figsize=(6,8))
sns.heatmap(num_cpus,annot=True,cmap="coolwarm")
plt.show()
```

```python
num_cpus=tips.corr(numeric_only=True)
plt.figure(figsize=(6,8))
sns.heatmap(num_cpus,annot=True,cmap="YlGnBu")
plt.show()
```
